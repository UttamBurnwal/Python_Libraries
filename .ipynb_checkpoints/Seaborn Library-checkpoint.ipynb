```python
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Seaborn has many datasets integrated in its library

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

### Line Plot

```python
sns.lineplot(x = 'timepoint', y = 'signal', data = fmri)
plt.show()
```

```python
# Highlight a particular column in the plot (e.g.: Event column in this case)

sns.lineplot(x = 'timepoint', y = 'signal', data = fmri, hue = 'event')
plt.show()
```

```python
# Adding style

sns.lineplot(x = 'timepoint', y = 'signal', data = fmri, hue = 'event', style = 'event')
plt.show()
```

```python
sns.lineplot(x = 'timepoint', y = 'signal', data = fmri, hue = 'event', style = 'region')
plt.show()
```

```python
# Adding markers

sns.lineplot(x = 'timepoint', y = 'signal', data = fmri, hue = 'event', style = 'event', markers = True)
plt.show()
```

### Bar Plot

### Scatterplot

```python
iris = pd.read_csv('C:\\Users\\Admin\\Downloads/iris.csv', index_col = 'Id')
```

```python
iris.head()
```

```python
iris.columns  = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
```

```python
iris.head()
```

```python
sns.scatterplot(x = 'sepal_length', y = 'petal_length', data = iris)
plt.show()
```

```python
# Hue

sns.scatterplot(x = 'sepal_length', y = 'petal_length', hue = 'species', data = iris)
plt.show()
```

```python
# Style

sns.scatterplot(x = 'sepal_length', y = 'petal_length', hue = 'species', style = 'species', data = iris)
plt.show()
```

```python
# Setting hue by a numerical column

sns.scatterplot(x = 'sepal_length', y = 'petal_length', hue = 'petal_length', data = iris)
plt.show()
```

```python
# hue = 'petal_length' & style = 'species'

sns.scatterplot(x = 'sepal_length', y = 'petal_length', hue = 'petal_length', style = 'species', data = iris)
plt.show()
```

### Dist Plot

```python
diamonds = pd.read_csv('C:/Users/Admin/Downloads/diamonds.csv', index_col = 'Unnamed: 0')
diamonds.head()
```

```python
sns.distplot(diamonds['price'])
plt.show()
```

```python
# Distribution without Histograms

sns.distplot(diamonds['price'], hist= False)
plt.show()
```

```python
# Adding color

sns.distplot(diamonds['price'], color = 'red')
plt.show()
```

```python
# Distribution plot without frequency curve: Only Histogram

sns.distplot(diamonds['price'], color = 'red', kde = False)
plt.show()
```

```python
# Changing Bins

sns.distplot(diamonds['price'], color = 'red', kde = False, bins = 10)
plt.show()
```

```python
# Plotting on a vertical axis

sns.distplot(diamonds['price'], color = 'orange', vertical = True)
plt.show()
```

### Joint Plot: Combination of a scatter plot and a histogram

```python
sns.jointplot(x = 'sepal_length', y = 'petal_length', data = iris)
plt.show()
```

```python
# Changing color

sns.jointplot(x = 'sepal_length', y = 'petal_length', data = iris, color = 'olive')
plt.show()
```

```python
# Adding a regression line on the scatterplot

sns.jointplot(x = 'sepal_length', y = 'petal_length', data = iris, kind = 'reg', color = 'red')
plt.show()
```

### Box Plot

```python
churn = pd.read_csv('C:\\Users\\Admin\\Downloads/customer_churn.csv')
churn.head()
```

```python
churn.columns
```

```python
sns.boxplot(x = 'Churn', y = 'tenure', data = churn)
plt.show()
```

```python
sns.boxplot(x = 'InternetService', y = 'MonthlyCharges', data = churn)
plt.show()
```

```python
# Color Palette

sns.boxplot(x = 'Contract', y = 'tenure', data = churn, palette ='Set1')
plt.show()
```

```python
# Changing the width

sns.boxplot(x = 'Contract', y = 'tenure', data = churn, palette ='Set1', linewidth = 3)
plt.show()
```

```python
# Changing the order of X -axis values

sns.boxplot(x = 'Contract', y = 'tenure', data = churn, palette ='Set1', linewidth = 3, order = ['Two year', 'Month-to-month', 'One year'])
plt.show()
```

```python
# Adding Hue

sns.boxplot(x = 'Contract', y = 'tenure', data = churn, palette ='Set1', hue = 'PaymentMethod')
plt.show()
```
