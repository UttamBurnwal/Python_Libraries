```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,11)
x
```

```python
y = x*2
y
```

## Line Plot

```python
plt.plot(x, y)
plt.show()
```

```python
# Adding Title, X-label and Y -label

plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('X-label')
plt.ylabel('y-label')
plt.show()
```

```python
# Changing colour

plt.plot(x, y, color = 'green')
plt.show()
```

```python
# Changing linestyle

plt.plot(x, y, linestyle = 'dashed')
plt.show()
```

```python
# Changing Width

plt.plot(x, y, linewidth = 5)
plt.show()
```

```python
# Marker Option
plt.plot(x, y, marker='o')
plt.show()
```

```python
x = np.arange(1,11)
```

```python
y1 = 2*x
y2 = 3*x
```

```python
# 2 line plots in same figure


plt.plot(x, y1, color='red', linestyle=':', linewidth=3, markersize=12)
plt.plot(x, y2, color = 'green', linestyle = '-.', linewidth = 3)
plt.title('Line Plot')
plt.xlabel('x_axis')
plt.ylabel('y_label')
plt.grid(True)
plt.show()
```

```python
# 2 sub-plots

plt.subplot(1, 2, 1)
plt.plot(x, y1, color = 'red', linestyle=':', linewidth=3, markersize=12)

plt.subplot(1, 2, 2)
plt.plot(x, y2, color = 'green', linestyle = '-.', linewidth = 3)
plt.show()
```

## Bar Plot

```python
student = {'Sam': 56, 'Patrick': 78, 'Jane': 84, 'Ronit': 76, 'Manny': 75}
```

```python
names = list(student.keys())
values = list(student.values())
```

```python
plt.bar(names, values)
plt.show()
```

```python
plt.bar(names, values)
plt.title('Marks of Students')
plt.xlabel('Names')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

```python
plt.bar(names, values, color = 'red', edgecolor = 'black')
plt.title('Marks of Students')
plt.xlabel('Names')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

```python
# Horizontal Bar plot (barh)

plt.barh(names, values, color = 'red', edgecolor = 'black')
plt.title('Marks of Students')
plt.xlabel('Names')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

## Scatter Plot

```python
# Same number of elements should be present on both X and Y list (x,y)
x = [1,2,5,6,9,4,8,3]
y = [45,32,41,37,23,17,10,98]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y, marker = '*')
plt.show()
```

```python
# C means color

plt.scatter(x,y, c = 'green')
plt.show()
```

```python
# s means size

plt.scatter(x,y, s = 100)
plt.show()
```

```python
plt.scatter(x,y, color = 'orange', s = 200, marker = '*')
plt.show()
```

```python
x = [1,2,5,6,9,4,8,3]
y = [45,32,41,37,23,17,10,98]
z = [23,43,12,37,49,14,17,42]
```

```python
plt.scatter(x, y, c='orange', s=100, marker = '*')
plt.scatter(x, z, c='green', s =100)
plt.show()
```

```python
# Sub-Plot

plt.subplot(1,2,1)
plt.scatter(x,y, c='orange', marker = '*', s = 100)

plt.subplot(1,2,2)
plt.scatter(x,z, c = 'green', s = 100)
plt.show()
```

## Pie Chart

```python
student = {'Sam': 56, 'Patrick': 78, 'Jane': 84, 'Ronit': 76, 'Manny': 75}
```

```python
names = list(student.keys())
values = list(student.values())
```

```python
plt.pie(values, labels = names)
plt.show()
```

```python
plt.pie(values, labels = names, autopct = '%0.2f%%', colors = ['green', 'red', 'orange', 'blue', 'yellow'], shadow = True)
plt.show()
```

# Dognut Chart

```python
plt.pie(values, labels = names, radius =2)
plt.pie([1],colors = 'w', radius =1)
plt.show()
```
